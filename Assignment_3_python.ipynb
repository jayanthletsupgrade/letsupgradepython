```
for num in range(1042000,702648265+1):
    sum=0
    order=len(str(num))
    temp=num 
    while temp>0:
        digit=temp%10
        sum+=digit**order
        temp//=10
    if num==sum:
       print(num," armstrong number")
       break
    else:
       print (num,"not Armstrong number")
```
